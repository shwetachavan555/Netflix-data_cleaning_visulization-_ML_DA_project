```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

```

```python
# Count the number of Movies and TV Shows
type_counts = data['type'].value_counts()
```

```python
# Plot the distribution
plt.figure(figsize=(8, 6))

sns.barplot(
    x=type_counts.index,
    y=type_counts.values,
    hue=type_counts.index,   # required with palette
    palette='Set2',
    legend=False
)

plt.title('Distribution of Content by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
#most common genere
data['genres'] = data['listed_in'].apply(lambda x: x.split(', '))
all_genres = sum(data['genres'], [])
genre_counts = pd.Series(all_genres).value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

freq = data['type'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# LEFT: Count plot
sns.countplot(data=data, x='type', ax=axes[0])
axes[0].set_title('Content Count')
axes[0].set_xlabel('')
axes[0].set_ylabel('Total')

# RIGHT: Pie chart
axes[1].pie(
    freq.values,
    labels=freq.index,
    autopct='%.0f%%',
    startangle=90
)
axes[1].set_title('Content Percentage')

# MAIN TITLE
plt.suptitle('Total Content on Netflix', fontsize=20)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Extract year, month from date_added (run once if not done)
data['year'] = data['date_added'].dt.year
data['month'] = data['date_added'].dt.month

# Monthly releases
monthly_movie_release = (
    data[data['type'] == 'Movie']['month']
    .value_counts()
    .sort_index()
)

monthly_series_release = (
    data[data['type'] == 'TV Show']['month']
    .value_counts()
    .sort_index()
)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_movie_release.index,
    monthly_movie_release.values,
    label='Movies',
    marker='o'
)
plt.plot(
    monthly_series_release.index,
    monthly_series_release.values,
    label='TV Shows',
    marker='o'
)

plt.xlabel("Months")
plt.ylabel("Frequency of Releases")
plt.xticks(
    range(1, 13),
    ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
     'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
)
plt.legend()
plt.grid(True)
plt.title("Monthly Releases of Movies and TV Shows on Netflix")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Define the missing variables
yearly_movie_releases = data[data['type'] == 'Movie']['release_year'].value_counts().sort_index()
yearly_series_releases = data[data['type'] == 'TV Show']['release_year'].value_counts().sort_index()

# 2. Now run your plotting code
plt.figure(figsize=(12, 6))

plt.plot(yearly_movie_releases.index, yearly_movie_releases.values, 
         label='Movies', linewidth=2, marker='o', markersize=4)
plt.plot(yearly_series_releases.index, yearly_series_releases.values, 
         label='TV Shows', linewidth=2, marker='o', markersize=4)

# 3. Formatting
plt.title("Yearly Releases of Movies and TV Shows on Netflix", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Filter for Movies and separate the comma-separated genres
movie_genres = data[data['type'] == 'Movie']['listed_in'].str.split(', ')

# 2. Explode the lists into individual rows and count the occurrences
exploded_movie_genres = movie_genres.explode().value_counts().head(10)

# 3. Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x=exploded_movie_genres.values, 
            y=exploded_movie_genres.index, 
            hue=exploded_movie_genres.index, 
            palette="viridis", 
            legend=False)

plt.title("Top 10 Most Popular Movie Genres on Netflix", fontsize=15)
plt.xlabel("Number of Movies", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()
```
